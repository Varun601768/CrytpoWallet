```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime
import requests
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
# User inputs
crypto_ticker = 'ETH-USD'  # Change to any ticker (e.g., 'ADA-USD', 'BTC-USD')
currency_code = 'INR'      # Change to any currency (e.g., 'EUR', 'JPY')
api_key = '24707e87e2d3276d6f435d6c'   # Get from exchangerate-api.com
prediction_days = 7        # Days to predict
sequence_length = 60       # Days for LSTM input sequence
```

```python
# Set date range (1 year for faster processing)
end = datetime(2025, 6, 14)  # Current date: June 14, 2025
start = datetime(end.year-15, end.month, end.day)
```

```python
# Function to fetch crypto data
def fetch_crypto_data(ticker, start, end):
    try:
        data = yf.download(ticker, start=start, end=end)
        if data.empty:
            print(f"No data found for {ticker}. Trying earliest available data.")
            data = yf.download(ticker, start='2010-01-01', end=end)
        return data
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        return None
```

```python
# Function to get exchange rate
def get_exchange_rate(from_currency, to_currency, api_key):
    url = f"https://v6.exchangerate-api.com/v6/{api_key}/latest/{from_currency}"
    try:
        response = requests.get(url)
        data = response.json()
        return data['conversion_rates'][to_currency]
    except Exception as e:
        print(f"Error fetching exchange rate: {e}")
        return 1  # No conversion if API fails
```

```python
# Fetch data
stock_data = fetch_crypto_data(crypto_ticker, start, end)
if stock_data is None:
    raise ValueError("Failed to fetch data. Exiting.")
```

```python
# Preprocessing: Convert price columns to target currency
exchange_rate = get_exchange_rate('USD', currency_code, api_key)
print(f"Exchange rate: 1 USD = {exchange_rate} {currency_code}")
price_columns = ['Open', 'High', 'Low', 'Close']
for col in price_columns:
    stock_data[col] = stock_data[col] * exchange_rate
stock_data = stock_data.dropna()  # Remove missing values
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Mock sentiment data (can be replaced with live API data later)
mock_posts = [
    "Ethereum to the moon! #ETH",
    "Bad news for ETH, regulations coming.",
    "Bullish on Ethereum, great staking yields!",
    "ETH price dip, time to buy.",
    "Worried about ETH market, too volatile.",
    "ETH is the future of DeFi!",
    "Market crash fears for Ethereum.",
    "ETH staking is amazing, buy now!"
]

# Analyze sentiment
analyzer = SentimentIntensityAnalyzer()
sentiments = [analyzer.polarity_scores(post)['compound'] for post in mock_posts]

# Use recent dates from stock_data to align sentiments
# Ensure stock_data exists and is already fetched before this step
if stock_data is None:
    raise ValueError("stock_data must be fetched before running sentiment analysis.")

# Match sentiment to most recent dates in stock_data
sentiment_dates = stock_data.index[-len(sentiments):] if len(sentiments) <= len(stock_data) else stock_data.index

# Create sentiment DataFrame
sentiment_df = pd.DataFrame({'Sentiment': sentiments[:len(sentiment_dates)]}, index=sentiment_dates)

# Ensure stock_data has single-level columns before join
if isinstance(stock_data.columns, pd.MultiIndex):
    stock_data.columns = ['_'.join(col).strip() for col in stock_data.columns.values]

# Join sentiment with stock_data
try:
    stock_data = stock_data.join(sentiment_df, how='left').fillna(0)  # Fill missing sentiment with neutral
    print("✅ Sentiment successfully joined with stock_data.")
except Exception as e:
    print(f"❌ Failed to join sentiment data: {e}")
```

```python
# Get latest price
if not stock_data.empty:
    # Adjust column name to match your actual structure
    close_col = [col for col in stock_data.columns if 'Close' in col][0]
    latest_price = stock_data[close_col].iloc[-1].item()
    print(f"Latest {crypto_ticker} price ({currency_code}): {latest_price:.2f}")
else:
    print(f"No data available for {crypto_ticker}.")
    raise ValueError("Cannot proceed with empty data.")
```

```python
from sklearn.preprocessing import MinMaxScaler

# 🔍 Step 1: Flatten MultiIndex column names if necessary
if isinstance(stock_data.columns, pd.MultiIndex):
    stock_data.columns = ['_'.join(col).strip() for col in stock_data.columns.values]

# 🔄 Step 2: Auto-strip ticker suffixes (e.g., 'Close_ETH-USD' ➝ 'Close')
stock_data.columns = [col.split('_')[0] if '_' in col else col for col in stock_data.columns]

# 🧼 Step 3: Ensure required features are present
required_features = ['Close', 'Volume', 'Sentiment']
missing = [feat for feat in required_features if feat not in stock_data.columns]
if missing:
    raise KeyError(f"Missing required columns: {missing}")

# 📊 Step 4: Extract and scale features
features = ['Close', 'Volume', 'Sentiment']
data = stock_data[features].values
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

print("Scaled data shape:", scaled_data.shape)
```

```python
# Create sequences for LSTM
X, y = [], []
for i in range(sequence_length, len(scaled_data) - prediction_days):
    X.append(scaled_data[i-sequence_length:i])
    y.append(scaled_data[i+prediction_days-1, 0])  # Predict 'Close' price
X, y = np.array(X), np.array(y)
```

```python
# Split data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dropout, Dense

model = Sequential([
    Input(shape=(X_train.shape[1], X_train.shape[2])),
    LSTM(50, return_sequences=True),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# --- Training with Early Stopping ---
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
history = model.fit(X_train, y_train, batch_size=32, epochs=100, 
                   validation_data=(X_test, y_test), callbacks=[early_stopping], verbose=1)
```

```python
# --- Evaluate Model on Test Set ---
y_pred = model.predict(X_test, verbose=0)
```

```python
# Inverse transform predictions and actual values
scaler = MinMaxScaler(feature_range=(0, 1)).fit(stock_data[features])  # Ensure scaler is fitted
y_pred = scaler.inverse_transform(np.hstack([y_pred, np.zeros((y_pred.shape[0], len(features)-1))]))[:, 0]
y_test_inv = scaler.inverse_transform(np.hstack([y_test.reshape(-1, 1), np.zeros((y_test.shape[0], len(features)-1))]))[:, 0]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Calculate metrics
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred))
mae = mean_absolute_error(y_test_inv, y_pred)

print(f"Test RMSE: {rmse:.2f} {currency_code}")
print(f"Test MAE: {mae:.2f} {currency_code}")
```

```python
import tweepy
import time
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import numpy as np
from datetime import datetime, timedelta

# Initialize client
client = tweepy.Client(bearer_token='AAAAAAAAAAAAAAAAAAAAADn%2B2QEAAAAAZ4vZ39M2GjFiUIyQFafho8oKRrI%3DSdsbDz6aAEjNIX8odYeXy84cV7CjSjkPz07xFVd5LI8xKmAOg3')

def fetch_tweets_with_backoff(query, max_results=50, max_retries=5):
    """
    Fetch tweets with exponential backoff retry strategy
    """
    base_wait_time = 60  # Start with 1 minute for rate limits
    
    for attempt in range(max_retries):
        try:
            tweets = client.search_recent_tweets(
                query=query, 
                max_results=max_results,
                tweet_fields=['created_at', 'public_metrics']  # Get additional data
            )
            print(f"Successfully fetched tweets on attempt {attempt + 1}")
            return tweets
            
        except tweepy.errors.TooManyRequests as e:
            # Check rate limit reset time from headers if available
            reset_time = getattr(e.response, 'headers', {}).get('x-rate-limit-reset')
            if reset_time:
                reset_timestamp = int(reset_time)
                wait_time = max(reset_timestamp - int(time.time()) + 10, base_wait_time)
                print(f"Rate limit hit. Waiting {wait_time} seconds until reset...")
            else:
                wait_time = base_wait_time * (2 ** attempt)  # Exponential backoff
                print(f"Rate limit hit. Waiting {wait_time} seconds... (Attempt {attempt + 1}/{max_retries})")
            
            if attempt < max_retries - 1:
                time.sleep(wait_time)
            
        except tweepy.errors.Unauthorized:
            print("Authentication error. Check your bearer token.")
            return None
            
        except Exception as e:
            print(f"Unexpected error: {e}")
            if attempt < max_retries - 1:
                time.sleep(base_wait_time)
            else:
                return None
    
    print("Max retries reached. Could not fetch tweets.")
    return None

def analyze_crypto_sentiment(crypto_ticker, sample_size=50):
    """
    Analyze sentiment for a cryptocurrency with improved error handling
    """
    # Clean the ticker symbol
    ticker_symbol = crypto_ticker.split('-')[0].upper()
    
    # Create a more comprehensive query
    query = f"({ticker_symbol} OR #{ticker_symbol}) -is:retweet -is:reply lang:en"
    
    print(f"Searching for tweets about: {ticker_symbol}")
    print(f"Query: {query}")
    
    # Fetch tweets
    tweets = fetch_tweets_with_backoff(query, max_results=sample_size)
    
    if not tweets or not tweets.data:
        print("No tweets found or error occurred.")
        return {
            'ticker': ticker_symbol,
            'tweet_count': 0,
            'avg_sentiment': 0,
            'sentiment_distribution': {'positive': 0, 'neutral': 0, 'negative': 0}
        }
    
    # Analyze sentiment
    analyzer = SentimentIntensityAnalyzer()
    sentiments = []
    sentiment_distribution = {'positive': 0, 'neutral': 0, 'negative': 0}
    
    print(f"Analyzing {len(tweets.data)} tweets...")
    
    for tweet in tweets.data:
        try:
            # Get compound score
            scores = analyzer.polarity_scores(tweet.text)
            compound_score = scores['compound']
            sentiments.append(compound_score)
            
            # Categorize sentiment
            if compound_score >= 0.05:
                sentiment_distribution['positive'] += 1
            elif compound_score <= -0.05:
                sentiment_distribution['negative'] += 1
            else:
                sentiment_distribution['neutral'] += 1
                
        except Exception as e:
            print(f"Error analyzing tweet: {e}")
            continue
    
    # Calculate results
    avg_sentiment = np.mean(sentiments) if sentiments else 0
    
    return {
        'ticker': ticker_symbol,
        'tweet_count': len(sentiments),
        'avg_sentiment': avg_sentiment,
        'sentiment_distribution': sentiment_distribution,
        'sentiment_scores': sentiments
    }

# Usage example
if __name__ == "__main__":
    # Example usage
    crypto_ticker = "BTC-USD"  # or whatever your crypto_ticker variable contains
    
    # Analyze sentiment
    results = analyze_crypto_sentiment(crypto_ticker)
    
    # Display results
    print("\n" + "="*50)
    print("SENTIMENT ANALYSIS RESULTS")
    print("="*50)
    print(f"Cryptocurrency: {results['ticker']}")
    print(f"Tweets analyzed: {results['tweet_count']}")
    print(f"Average sentiment: {results['avg_sentiment']:.3f}")
    print(f"Sentiment distribution:")
    for sentiment, count in results['sentiment_distribution'].items():
        percentage = (count / results['tweet_count'] * 100) if results['tweet_count'] > 0 else 0
        print(f"  {sentiment.capitalize()}: {count} ({percentage:.1f}%)")
    
    # Interpret sentiment
    if results['avg_sentiment'] > 0.05:
        print("Overall sentiment: POSITIVE 📈")
    elif results['avg_sentiment'] < -0.05:
        print("Overall sentiment: NEGATIVE 📉")
    else:
        print("Overall sentiment: NEUTRAL ➡️")
```
