```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime
import requests
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
# User inputs
crypto_ticker = 'ETH-USD'  # Change to any ticker (e.g., 'ADA-USD', 'BTC-USD')
currency_code = 'INR'      # Change to any currency (e.g., 'EUR', 'JPY')
api_key = '24707e87e2d3276d6f435d6c'   # Get from exchangerate-api.com
prediction_days = 7        # Days to predict
sequence_length = 60       # Days for LSTM input sequence
```

```python
# Set date range (1 year for faster processing)
end = datetime(2025, 6, 14)  # Current date: June 14, 2025
start = datetime(end.year-15, end.month, end.day)
```

```python
# Function to fetch crypto data
def fetch_crypto_data(ticker, start, end):
    try:
        data = yf.download(ticker, start=start, end=end)
        if data.empty:
            print(f"No data found for {ticker}. Trying earliest available data.")
            data = yf.download(ticker, start='2010-01-01', end=end)
        return data
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        return None
```

```python
# Function to get exchange rate
def get_exchange_rate(from_currency, to_currency, api_key):
    url = f"https://v6.exchangerate-api.com/v6/{api_key}/latest/{from_currency}"
    try:
        response = requests.get(url)
        data = response.json()
        return data['conversion_rates'][to_currency]
    except Exception as e:
        print(f"Error fetching exchange rate: {e}")
        return 1  # No conversion if API fails
```

```python
# Fetch data
stock_data = fetch_crypto_data(crypto_ticker, start, end)
if stock_data is None:
    raise ValueError("Failed to fetch data. Exiting.")
```

```python
# Preprocessing: Convert price columns to target currency
exchange_rate = get_exchange_rate('USD', currency_code, api_key)
print(f"Exchange rate: 1 USD = {exchange_rate} {currency_code}")
price_columns = ['Open', 'High', 'Low', 'Close']
for col in price_columns:
    stock_data[col] = stock_data[col] * exchange_rate
stock_data = stock_data.dropna()  # Remove missing values
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Mock sentiment data (can be replaced with live API data later)
mock_posts = [
    "Ethereum to the moon! #ETH",
    "Bad news for ETH, regulations coming.",
    "Bullish on Ethereum, great staking yields!",
    "ETH price dip, time to buy.",
    "Worried about ETH market, too volatile.",
    "ETH is the future of DeFi!",
    "Market crash fears for Ethereum.",
    "ETH staking is amazing, buy now!"
]

# Analyze sentiment
analyzer = SentimentIntensityAnalyzer()
sentiments = [analyzer.polarity_scores(post)['compound'] for post in mock_posts]

# Use recent dates from stock_data to align sentiments
# Ensure stock_data exists and is already fetched before this step
if stock_data is None:
    raise ValueError("stock_data must be fetched before running sentiment analysis.")

# Match sentiment to most recent dates in stock_data
sentiment_dates = stock_data.index[-len(sentiments):] if len(sentiments) <= len(stock_data) else stock_data.index

# Create sentiment DataFrame
sentiment_df = pd.DataFrame({'Sentiment': sentiments[:len(sentiment_dates)]}, index=sentiment_dates)

# Ensure stock_data has single-level columns before join
if isinstance(stock_data.columns, pd.MultiIndex):
    stock_data.columns = ['_'.join(col).strip() for col in stock_data.columns.values]

# Join sentiment with stock_data
try:
    stock_data = stock_data.join(sentiment_df, how='left').fillna(0)  # Fill missing sentiment with neutral
    print("âœ… Sentiment successfully joined with stock_data.")
except Exception as e:
    print(f"âŒ Failed to join sentiment data: {e}")
```

```python
# Get latest price
if not stock_data.empty:
    # Adjust column name to match your actual structure
    close_col = [col for col in stock_data.columns if 'Close' in col][0]
    latest_price = stock_data[close_col].iloc[-1].item()
    print(f"Latest {crypto_ticker} price ({currency_code}): {latest_price:.2f}")
else:
    print(f"No data available for {crypto_ticker}.")
    raise ValueError("Cannot proceed with empty data.")
```

```python
from sklearn.preprocessing import MinMaxScaler

# ðŸ” Step 1: Flatten MultiIndex column names if necessary
if isinstance(stock_data.columns, pd.MultiIndex):
    stock_data.columns = ['_'.join(col).strip() for col in stock_data.columns.values]

# ðŸ”„ Step 2: Auto-strip ticker suffixes (e.g., 'Close_ETH-USD' âž 'Close')
stock_data.columns = [col.split('_')[0] if '_' in col else col for col in stock_data.columns]

# ðŸ§¼ Step 3: Ensure required features are present
required_features = ['Close', 'Volume', 'Sentiment']
missing = [feat for feat in required_features if feat not in stock_data.columns]
if missing:
    raise KeyError(f"Missing required columns: {missing}")

# ðŸ“Š Step 4: Extract and scale features
features = ['Close', 'Volume', 'Sentiment']
data = stock_data[features].values
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

print("Scaled data shape:", scaled_data.shape)
```

```python
# Create sequences for LSTM
X, y = [], []
for i in range(sequence_length, len(scaled_data) - prediction_days):
    X.append(scaled_data[i-sequence_length:i])
    y.append(scaled_data[i+prediction_days-1, 0])  # Predict 'Close' price
X, y = np.array(X), np.array(y)
```

```python
# Split data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dropout, Dense

model = Sequential([
    Input(shape=(X_train.shape[1], X_train.shape[2])),
    LSTM(50, return_sequences=True),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# --- Training with Early Stopping ---
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
history = model.fit(X_train, y_train, batch_size=32, epochs=100, 
                   validation_data=(X_test, y_test), callbacks=[early_stopping], verbose=1)
```

```python
# --- Evaluate Model on Test Set ---
y_pred = model.predict(X_test, verbose=0)
```

```python
# Inverse transform predictions and actual values
scaler = MinMaxScaler(feature_range=(0, 1)).fit(stock_data[features])  # Ensure scaler is fitted
y_pred = scaler.inverse_transform(np.hstack([y_pred, np.zeros((y_pred.shape[0], len(features)-1))]))[:, 0]
y_test_inv = scaler.inverse_transform(np.hstack([y_test.reshape(-1, 1), np.zeros((y_test.shape[0], len(features)-1))]))[:, 0]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Calculate metrics
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred))
mae = mean_absolute_error(y_test_inv, y_pred)

print(f"Test RMSE: {rmse:.2f} {currency_code}")
print(f"Test MAE: {mae:.2f} {currency_code}")
```

```python
# Plot test predictions (Matplotlib for simplicity)
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 6))
plt.plot(y_test_inv, label='Actual Prices', color='#1f77b4')
plt.plot(y_pred, label='Predicted Prices', color='#ff7f0e', linestyle='--')
plt.title(f"{crypto_ticker} Test Set Predictions ({currency_code})")
plt.xlabel("Time")
plt.ylabel(f"Price ({currency_code})")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# --- Predict Future Prices (7 Days) ---
last_sequence = scaled_data[-sequence_length:]
future_predictions = []
current_sequence = last_sequence.reshape(1, sequence_length, len(features))
for _ in range(prediction_days):
    try:
        pred = model.predict(current_sequence, verbose=0)[0, 0]
        future_predictions.append(pred)
        avg_sentiment = np.mean(scaled_data[-5:, -1])  # Use average of last 5 days' sentiment
        new_row = np.array([[pred, current_sequence[0, -1, 1], avg_sentiment]])
        current_sequence = np.append(current_sequence[:, 1:, :], new_row.reshape(1, 1, len(features)), axis=1)
    except Exception as e:
        print(f"Error during prediction: {e}")
        break
```

```python
# Inverse scale predictions
future_predictions = np.array(future_predictions).reshape(-1, 1)
dummy_array = np.zeros((len(future_predictions), len(features)))
dummy_array[:, 0] = future_predictions[:, 0]
future_predictions = scaler.inverse_transform(dummy_array)[:, 0]
```

```python
import numpy as np
import pandas as pd
import tweepy
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import os

# Ensure features match the model's training data
features = ['Close', 'Volume', 'Sentiment']
scaler = MinMaxScaler(feature_range=(0, 1)).fit(stock_data[features])
scaled_data = scaler.transform(stock_data[features])

# Check for cached sentiment
cache_file = 'sentiment_cache.csv'
def load_cached_sentiment():
    if os.path.exists(cache_file):
        try:
            return pd.read_csv(cache_file)['sentiment'].iloc[-1]
        except Exception as e:
            print(f"Error reading cache: {e}")
    return None

# Function to save sentiment to cache
def save_sentiment_to_cache(sentiment):
    try:
        pd.DataFrame({'sentiment': [sentiment], 'timestamp': [pd.Timestamp.now()]}).to_csv(cache_file, index=False)
    except Exception as e:
        print(f"Error saving to cache: {e}")

# Function to check API rate limit status
def check_rate_limit(client):
    try:
        rate_limit = client.get_rate_limit_status()
        search_limit = rate_limit['resources']['search']['/search/tweets']['remaining']
        return search_limit > 0
    except tweepy.TweepyException as e:
        print(f"Error checking rate limit: {e}")
        return False

# Function to fetch real-time sentiment
def get_real_time_sentiment(crypto_ticker):
    # Check cache first
    cached_sentiment = load_cached_sentiment()
    if cached_sentiment is not None:
        print("Using cached sentiment score.")
        return cached_sentiment

    # Initialize Twitter/X API client
    try:
        client = tweepy.Client(bearer_token='AAAAAAAAAAAAAAAAAAAAAP7k2QEAAAAAgUt7ENZ6Y3U9KpQmZJAZfgfSVqw%3D7zjYndflt0Cu7AKbCeBooNhRTe2ZYQpgJQ5e0IreB6lXufRDLK')  # Replace with your actual bearer token
    except tweepy.TweepyException as e:
        print(f"Error initializing Twitter/X API: {e}")
        return np.mean(scaled_data[-5:, -1]) if len(scaled_data) >= 5 else 0

    # Check rate limit before making API call
    if not check_rate_limit(client):
        print("Rate limit or monthly cap exceeded. Using historical sentiment.")
        return np.mean(scaled_data[-5:, -1]) if len(scaled_data) >= 5 else 0

    try:
        query = f"{crypto_ticker.split('-')[0]} -is:retweet lang:en"
        tweets = client.search_recent_tweets(query=query, max_results=100)
        analyzer = SentimentIntensityAnalyzer()
        if tweets.data:
            sentiments = [analyzer.polarity_scores(tweet.text)['compound'] for tweet in tweets.data]
            avg_sentiment = np.mean(sentiments)
            save_sentiment_to_cache(avg_sentiment)  # Cache the new sentiment
            return avg_sentiment
        else:
            print("No tweets found for query.")
            return np.mean(scaled_data[-5:, -1]) if len(scaled_data) >= 5 else 0
    except tweepy.TweepyException as e:
        print(f"Error fetching sentiment: {e}")
        return np.mean(scaled_data[-5:, -1]) if len(scaled_data) >= 5 else 0

# Fetch or load sentiment
avg_sentiment = get_real_time_sentiment(crypto_ticker)
print(f"Using sentiment score: {avg_sentiment}")

# Predict future prices (7 days) with cached sentiment
last_sequence = scaled_data[-sequence_length:]
future_predictions = []
current_sequence = last_sequence.reshape(1, sequence_length, len(features))
for _ in range(prediction_days):
    try:
        pred = model.predict(current_sequence, verbose=0)[0, 0]
        future_predictions.append(pred)
        new_row = np.array([[pred, current_sequence[0, -1, 1], avg_sentiment]])
        current_sequence = np.append(current_sequence[:, 1:, :], new_row.reshape(1, 1, len(features)), axis=1)
    except Exception as e:
        print(f"Error during prediction: {e}")
        break

# Check if predictions were generated
if len(future_predictions) == 0:
    print("No predictions generated. Check model input dimensions or data.")
else:
    # Inverse scale predictions
    future_predictions = np.array(future_predictions).reshape(-1, 1)
    dummy_array = np.zeros((len(future_predictions), len(features)))
    dummy_array[:, 0] = future_predictions[:, 0]
    future_predictions = scaler.inverse_transform(dummy_array)[:, 0]

    # Define future dates
    future_dates = pd.date_range(start=stock_data.index[-1] + pd.Timedelta(days=1), periods=len(future_predictions)).tolist()

    # Display predicted prices
    print(f"Predicted {crypto_ticker} prices for next {len(future_predictions)} days ({currency_code}):")
    for i, price in enumerate(future_predictions, 1):
        print(f"Day {i}: {price:.2f} {currency_code}")
```
